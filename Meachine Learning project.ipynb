```python
# Group by Category and calculate total sales
sales_by_category = sss.groupby('Category')['Sales'].sum().sort_values()
```

```python
# Grouping data by region to calculate sales and profit contributions
region_summary = sss.groupby('Region')[['Sales', 'Profit']].sum()

# Pie chart for Sales and Profit by Region
fig, axes = plt.subplots(1, 2, figsize=(8, 7))

# Sales contribution pie chart
axes[0].pie(region_summary['Sales'], labels=region_summary.index, autopct='%1.1f%%', startangle=140)
axes[0].set_title('Sales Contribution by Region')

# Profit contribution pie chart
axes[1].pie(region_summary['Profit'], labels=region_summary.index, autopct='%1.1f%%', startangle=140)
axes[1].set_title('Profit Contribution by Region')

plt.show()
```

```python
# naive bias 
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model.fit(xtr,ytr)                                 
ypr = model.predict(xts)
accuracies.append(accuracy_score(yts,ypr)*100)
accuracies
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
for i in range(2,11):
    model= KNeighborsClassifier(n_neighbors=i)
    model.fit(xtr,ytr)                                 # Best Accuracy at 9th iteration= 81.73982762076568
    ypr = model.predict(xts)
    print(f' the accuracy of KNN with {i} is {accuracy_score(yts,ypr)*100}')
```

```python
model= KNeighborsClassifier(n_neighbors=9)
model.fit(xtr,ytr)
ypr = model.predict(xts)
accuracies.append(accuracy_score(yts,ypr)*100)
accuracies
```

```python
# applying RandomForest
from sklearn.ensemble import RandomForestClassifier
for i in range(2,11):
    model= RandomForestClassifier(max_depth=i)
    model.fit(xtr,ytr)
    ypr = model.predict(xts)
    print(f' the accuracy of Random forest with {i} is {accuracy_score(yts,ypr)*100}')
```

```python
model= RandomForestClassifier(max_depth=9)
model.fit(xtr,ytr)
ypr = model.predict(xts)
accuracies.append(accuracy_score(yts,ypr)*100)
accuracies
```
